```python
#source: http://koldunov.net/?p=920
#preambula
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data = pd.read_table('Kuzedeevo_data/temp_prec.txt', sep=';', \
                    parse_dates={'dates':[1, 2, 3]}, index_col = 0, \
                    header=None)
```

```python
data.head()
```

```python
data = data[[5, 7, 9, 11]]
data.head()
```

```python
data.columns = ['TMIN', 'TMEAN', 'TMAX', 'PRECIP']
```

```python
data.TMIN.plot()
```

```python
#if we have some missing values
#data = data.replace('     ', np.nan, regex=True)
```

```python
#import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

```python
data['month'] = data.index.month
data.head(2)
data.boxplot(column=['TMEAN'], by='month')
```

```python
data_year = data.resample('A', how = 'mean')
data_year.head()
data_year.TMEAN.plot()
plt.show()
```

```python
data_month = data.resample('M', how = 'mean')
data_month.head()
data_month.TMEAN.plot(); plt.show()
```
